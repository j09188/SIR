```
# 隔離模型 0~180天 模型擬合
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from lmfit import minimize, Parameters, Parameter, report_fit
from scipy.integrate import odeint

def f(y, t, paras):
    S, C, I, H, R, D, QC, QI, N = y
    try:
        alpha_1 = paras['alpha_1'].value
        alpha_2 = paras['alpha_2'].value
        beta = paras['beta'].value
        r_I = paras['r_I'].value
        r_H = paras['r_H'].value
        eta = paras['eta'].value
        omega = paras['omega'].value
        rho = paras['rho'].value
        tau = paras['tau'].value
        mu = paras['mu'].value
    except KeyError:
        alpha_1, alpha_2, beta, r_I, r_H, eta, omega, rho, tau, mu = paras
    # the model equations

    if (t//1) < 19.1:
      tau = 0
      mu = 0
      alpha = alpha_1
    else:
      alpha = alpha_2

    dSdt = - alpha*S*I/N + omega*C + mu*QC
    dCdt = alpha*S*I/N - omega*C - (alpha*S/N)*tau*(1-rho)*(beta*(I+QI)) - rho*C
    dIdt = rho*C - beta*I - r_I*I - (alpha*S/N)*tau*rho*(beta*I+beta*QI)
    dHdt = beta * I - r_H * H - eta * H + beta*QI
    dRdt = r_I*I + r_H*H + r_I*QI
    dDdt = eta * H
    dQCdt = (alpha*S/N)*tau*(1-rho)*(beta*(I+QI)) - mu*QC - rho*QC
    dQIdt = (alpha*S/N)*tau*rho*(beta*I+beta*QI) - r_I*QI - beta*QI + rho*QC
    dNdt = dSdt + dCdt + dIdt + dRdt
    return [dSdt, dCdt, dIdt, dHdt, dRdt, dDdt, dQCdt, dQIdt, dNdt]

def g(t, x0, paras):
    x = odeint(f, x0, t, args=(paras,))
    return x

def residual(paras, t, data, data2, data3):

    x0 = paras['S0'].value, paras['C0'].value, paras['I0'].value, paras['H0'].value, paras['R0'].value, paras['D0'].value, paras['QC0'].value, paras['QI0'].value, paras['N0'].value
    model = g(t, x0, paras)

    # you only have data for one of your variables
    x2_model = model[:, 3] + model[:, 4] + model[:, 5]
    x3_model = model[:, 5]
    x4_model = model[:, 3]
    result = np.zeros(len(x2_model))
    for i in range(len(x2_model)):
        result[i]=(abs(x2_model[i] - data[i])**(2)) + (abs(x3_model[i] - data2[i])**(2)) + (abs(x4_model[i] - data3[i])**(2))
    return result.ravel()

# initial conditions
N0 = 60360000
S0 = N0-16
C0 = 10
I0 = 3
H0 = 3
R0 = 0
D0 = 0
QC0 = 0
QI0 = 0
y0 = [S0, C0, I0, H0, R0, D0, QC0, QI0, N0]
   
# measured data
x2_measured = Italy[:180] 
x3_measured = Italy_death[:180]
x4_measured = Italy_hospital[:180]
t_measured = np.linspace(0, len(x2_measured)-1, len(x2_measured))

# set parameters including bounds; you can also fix parameters (use vary=False)
params = Parameters()
params.add('S0', value=S0, vary=False)
params.add('C0', value=C0, vary=False)
params.add('I0', value=I0, vary=False)
params.add('H0', value=H0, vary=False)
params.add('R0', value=R0, vary=False)
params.add('D0', value=D0, vary=False)
params.add('QC0', value=QC0, vary=False)
params.add('QI0', value=QI0, vary=False)
params.add('N0', value=N0, vary=False)
params.add('alpha_1', value=0.5, min=0, max=3) # value=10, min=0, max=20
params.add('alpha_2', value=0.5, min=0, max=3)
params.add('beta', value=0.5, min=0, max=3)
params.add('r_H', value=0.1, min=0, max=1)
params.add('r_I', value=0.1, min=0, max=1)
params.add('eta', value=0.01, min=0, max=0.3)
params.add('omega', value=0.5, min=0, max=1)
params.add('rho', value=0.5, min=0, max=1)
params.add('tau', value=0.1, min=0, max=5)
params.add('mu', value=0.1, min=0, max=1)

# fit model
result = minimize(residual, params, args=(t_measured, x2_measured, x3_measured, x4_measured), method='Nelder-Mead', options={'maxiter':1e10})  
# check results of the fit
data_fitted = g(np.linspace(0, len(x2_measured)-1, len(x2_measured)), y0, result.params)

# plot fitted data
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 3] + data_fitted[:, 4] + data_fitted[:, 5], '-', linewidth=2, color='red', label='Total Infected (fit)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), x2_measured, '.', linewidth=2, color='orange', label='Total Infected (real)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 5], '-', linewidth=2, color='yellow', label='Death (fit)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), x3_measured, '.', linewidth=2, color='green', label='Death (real)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 3], '-', linewidth=2, color='blue', label='Hospitalized (fit)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), x4_measured, '.', linewidth=2, color='purple', label='Hospitalized (real)')
plt.legend()
f = plt.figure(figsize=(12,4))
ax1 = f.add_subplot(121)
ax1.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 1], 'g--', alpha=0.7, linewidth=2, label='C')
ax1.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 2], 'b--', alpha=0.7, linewidth=2, label='I')
ax1.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 7], 'y--', alpha=0.7, linewidth=2, label='QI')
plt.legend()
ax2 = f.add_subplot(122)
ax2.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 6], 'b--', alpha=0.7, linewidth=2, label='QC')
plt.legend()
# display fitted statistics
report_fit(result)

plt.show()
print(data_fitted[-1,])
```

```
#義大利 封城模型 180~300天
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from lmfit import minimize, Parameters, Parameter, report_fit
from scipy.integrate import odeint

def f(y, t, paras):
    S, I, H, R, D, QS, QI, QR, N = y
    try:
        alpha = paras['alpha'].value
        beta = paras['beta'].value
        r_I = paras['r_I'].value
        r_H = paras['r_H'].value
        eta = paras['eta'].value
        tau = paras['tau'].value
        mu = paras['mu'].value
    except KeyError:
        alpha, beta, r_I, r_H, eta, tau, mu = paras
    # the model equations

    dSdt = - alpha * S * I / N - S*tau*(beta*I+beta*QI) + mu*QS
    dIdt = alpha * S * I / N - beta * I - r_I *I - I*tau*(beta*I+beta*QI)
    dHdt = beta * I - r_H * H - eta * H + beta*QI
    dRdt = r_I *I + r_H * H - R*tau*(beta*I+beta*QI) + mu*QR
    dDdt = eta * H
    dQSdt = S*tau*(beta*I+beta*QI) - mu*QS
    dQIdt = I*tau*(beta*I+beta*QI) - r_I*QI - beta*QI 
    dQRdt = R*tau*(beta*I+beta*QI) + r_I*QI - mu*QR
    dNdt = dSdt + dIdt + dRdt
    return [dSdt, dIdt, dHdt, dRdt, dDdt, dQSdt, dQIdt, dQRdt, dNdt]


def g(t, x0, paras):
    x = odeint(f, x0, t, args=(paras,))
    return x

def residual(paras, t, data, data2, data3):

    x0 = paras['S0'].value, paras['I0'].value, paras['H0'].value, paras['R0'].value, paras['D0'].value, paras['QS0'].value, paras['QI0'].value, paras['QR0'].value, paras['N0'].value
    model = g(t, x0, paras)

    # you only have data for one of your variables
    x2_model = model[:, 2] + model[:, 3] + model[:, 4] + model[:, 7]
    x3_model = model[:, 4]
    x4_model = model[:, 2]
    result = np.zeros(len(x2_model))
    for i in range(len(x2_model)):
        result[i]=(abs(x2_model[i] - data[i])**(2)) + 10*(abs(x3_model[i] - data2[i])**(2)) + 10*(abs(x4_model[i] - data3[i])**(2))
    return result.ravel()

# initial conditions
N0 = 6.03210555e+07
S0 = 6.01091393e+07
I0 = 2.29993415e+02
H0 = 3.69725441e+02
R0 = 2.11686234e+05
D0 = 3.55747420e+04
QS0 = 0
QI0 = 0
QR0 = 0
y0 = [S0, I0, H0, R0, D0, QS0, QI0, QR0, N0]
   
# measured data
x2_measured = Italy[180:300]
x3_measured = Italy_death[180:300]
x4_measured = Italy_hospital[180:300]
t_measured = np.linspace(0, len(x2_measured)-1, len(x2_measured))

# set parameters including bounds; you can also fix parameters (use vary=False)
params = Parameters()
params.add('S0', value=S0, vary=False)
params.add('I0', value=I0, vary=False)
params.add('H0', value=H0, vary=False)
params.add('R0', value=R0, vary=False)
params.add('D0', value=D0, vary=False)
params.add('QS0', value=QS0, vary=False)
params.add('QI0', value=QI0, vary=False)
params.add('QR0', value=QR0, vary=False)
params.add('N0', value=N0, vary=False)
params.add('alpha', value=0.1, min=0, max=1) 
params.add('beta', value=0.1, min=0, max=1) 
params.add('r_I', value=0.01, min=0, max=1)
params.add('r_H', value=0.01, min=0, max=1)
params.add('eta', value=0.01, min=0, max=1) 
params.add('tau', value=1e-8, min=1e-12, max=1e-4) 
params.add('mu', value=0.1, min=1e-12, max=1) 

# fit model
result = minimize(residual, params, args=(t_measured, x2_measured, x3_measured, x4_measured), method='Nelder-Mead', options={'maxiter':1e10}, nan_policy='omit')  
# check results of the fit
data_fitted = g(np.linspace(0, len(x2_measured)-1, len(x2_measured)), y0, result.params)

# plot fitted data
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 2] + data_fitted[:, 3] + data_fitted[:, 4] + data_fitted[:, 7], '-', linewidth=2, color='red', label='Total Infected (fit)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), x2_measured, '.', linewidth=2, color='orange', label='Total Infected (real)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 4], '-', linewidth=2, color='yellow', label='Death (fit)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), x3_measured, '.', linewidth=2, color='green', label='Death (real)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 2], '-', linewidth=2, color='blue', label='Hospitalized (fit)')
plt.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), x4_measured, '.', linewidth=2, color='purple', label='Hospitalized (real)')
plt.legend()
f = plt.figure(figsize=(12,4))
ax1 = f.add_subplot(121)
ax1.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 1], 'b--', alpha=0.7, linewidth=2, label='I')
ax1.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 6], 'y--', alpha=0.7, linewidth=2, label='QI')
ax1.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 2], 'c--', alpha=0.7, linewidth=2, label='H')
ax2 = f.add_subplot(122)
ax2.plot(np.linspace(0, len(x2_measured)-1, len(x2_measured)), data_fitted[:, 5], 'b--', alpha=0.7, linewidth=2, label='QS')

plt.legend()
# display fitted statistics
report_fit(result)

plt.show()
print(data_fitted[-1,])
```
